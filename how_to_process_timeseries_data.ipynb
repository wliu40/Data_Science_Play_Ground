#### Put all to one function

```python
def process_data(arr, window_size, batch_size):
    """
    arr is a numpy array
    
    """
    x = tf.data.Dataset.from_tensor_slices(arr)
    x = x.window(window_size+1, shift=1, drop_remainder=True)
    x = x.flat_map(lambda w: w.batch(window_size+1))
    x = x.map(lambda w: (w[:-1], w[-1]))
    x = x.shuffle(buffer_size=10)
    x = x.batch(batch_size).prefetch(2)
    return x

arr = range(100)
x = process_data(arr, 10, 3)

for i,j in x:
    print(i.numpy())
    print(j.numpy())
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    # YOUR CODE HERE
    #series = tf.expand_dims(series, axis=-1)
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    ds = ds.shuffle(shuffle_buffer)
    ds = ds.map(lambda w: (w[:-1], w[-1]))
    return ds.batch(batch_size).prefetch(1)

arr = range(100)
x = windowed_dataset(arr, 10, 2, 10)

for i,j in x:
    print(i.numpy())
    print(j.numpy())
    break
```
